```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
# baca file .csv yang akan digunakan
# Sumber dataset: https://www.kaggle.com/harshitshankhdhar/imdb-dataset-of-top-1000-movies-and-tv-shows
df = pd.read_csv("D:\St.5\ML\Dataset\Tugas01-imdbtop1000.csv")
df.head
```

```python
# ambil nama film dan genrenya
df = df.loc[:, ['Series_Title', 'Genre']]
df.head()
```

```python
# ambil isi dari genre kemudian masukkan ke dalam list bernama data
data = list(df["Genre"].apply(lambda x:x.split(",")))

# ambil 10 data
data[:10]
```

```python
# akan digunakan algoritma apriori untuk mengubah nilai True/False menjadi 1/0.
# Genre yang digunakan dalam satu film akan diwakili oleh nilai 1.
# Genre yang tidak digunakan diwakili oleh nilai 0.
te = TransactionEncoder()
te_data = te.fit(data).transform(data)
df = pd.DataFrame(te_data, columns = te.columns_)
df = df.replace(False, 0)
df = df.replace(True, 1)
df
```

```python
# mencari frequent value dari genre menggunakan algoritma apriori
df = apriori(df, min_support = 0.1, use_colnames = True, verbose = 1)
df
```

```python
# akan dipilih nilai confidence sebesar 0.6 (60%).
# artinya saat genre A digunakan, dapat dikatakan bahwa persentase genre B digunakankan sebesar 60%
df_rules = association_rules(df, metric = "confidence", min_threshold = 0.6)
df_rules
```
